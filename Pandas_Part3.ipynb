GroupBy using pandas

Q. Why we need to perform groupBy
-  Whenever in dataset we have a catagorical column then wee need to perform groupby in a depth analysis


```python
# to perform groupby we need DataFrame
# syntax: df.groupby(ColumnName) --> groupby object
p.groupby('species')
```

```python
# to check actual values just perform typecasting
list(p.groupby('species'))  # --> (catagoty-dataframe)  it follows a-z order
```

```python
for i,j in p.groupby('species'):
  print(i)
```

```python
print(j)
```

```python
# fetch only Gentoo data
for i,j in p.groupby('species'):
  if i == 'Gentoo':
    print(j)
```

```python
# after groupby if we want a dataframe of specific catagory
p.groupby('species').get_group('Gentoo')
```

```python
p.groupby('species').get_group('Adelie')
```

```python
#  if we want first 3 columns
p.groupby('species').get_group('Adelie').iloc[:,:2]
```

```python
# using loc
p.groupby('species').get_group('Adelie').loc[:,:'bill_length_mm']
```

```python
# i want to retrive avarage body mass of each penguin as per catagory
p.groupby('species')['body_mass_g'].mean()
```

```python
p.groupby('species')['body_mass_g'].mean().plot(kind='bar',color=['r','y','g'])
```

```python
# island: mean() max() min()
p.groupby('island')['body_mass_g'].agg([max])
```

```python
import numpy as np
p.groupby('island')['body_mass_g'].agg([max,min,np.mean])
```

```python
p.groupby('island')['body_mass_g','bill_length_mm'].agg([max,min,np.mean])
```

```python
p.groupby('island')[['body_mass_g','bill_length_mm']].agg([max,min,np.mean])   ## duble [[]] removes warning
```

```python
# we can add more catagory in groupby
p.groupby(['island','species'])[['body_mass_g','bill_length_mm']].agg([max,min,np.mean])
```
