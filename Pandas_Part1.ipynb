#### loc and iloc for feature selection


```python
# loc: label(column_name) based selection [row_start:row_stop,column_start:column_stop]
data.loc[100:,:]   # start reading of the data from 100th index
```

```python
data.loc[100:111,:]   # start reading of the data from 100th index stop at 111
```

```python
# [: works for rows, : works for columns]
data.loc[:,:]
```

```python
data.loc[:,'total_rooms':'households']
```
